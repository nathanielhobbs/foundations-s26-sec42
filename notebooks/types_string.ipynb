### indexing

Indexing is when you get the value of a single item in a collection by means of its **index**. For strings, the collection of characters is stored as an array. So, indexes start at 0 and go up by 1 for each character.

```python
x = 'hello'

# indexing
print(x[0]) # print the value of the item at index 0
print(x[1]) # print the value of the item at index 1
print(x[2])
print(x[3])
print(x[4])
print(x[5]) # you can go too far
```

#### indices can be negative

```python
x = 'hello'

print(x[-1])
print(x[-2])
print(x[-3])
print(x[-4])
print(x[-5])
print(x[-6]) # too far
```

#### slicing (indexing many values)

The triplet start:stop:step defines a numbered sequence.

- start is **inclusive** and has a default of 0
- stop is **exclusive** and has a default of the end of the string
- step has a default of 1



```python
x = 'hello'

print(x[2:5:1]) # this makes the sequence: 2,3,4
print(x[2:5:]) # this uses the default step size
print(x[2:5]) # in fact, you can omit the step size all together
print(x[2:]) # this uses the default stop and step and is the most elegant here

```

```python
x = 'hello'

print(x[-1:-len(x)-1:-1]) # when stepping negative, default start/stop change
print(x[::-1]) # using the default values we can be very brief
```

```python
my_string = 'hello'
print(my_string[-1])
print(my_string[len(my_string)-1])
```
